### Extracting text from a website:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.healthline.com/nutrition/50-super-healthy-foods'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
text = soup.get_text()
print(text)
```

### Extracting links from a website:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.healthline.com/nutrition/50-super-healthy-foods'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
links = [link.get('href') for link in soup.find_all('a')]
print(links)
```

### Extracting data from a table on a website:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.healthline.com/nutrition/50-super-healthy-foods'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table')
headers = [header.get_text() for header in table.find_all('th')]
rows = []
for row in table.find_all('tr'):
    rows.append([data.get_text() for data in row.find_all('td')])
print(headers)
print(rows)
```
